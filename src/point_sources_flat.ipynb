```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining point sources
px = np.linspace(-5, 5, 20)
py = np.linspace(-5, 5, 20)
pz = 0  # sources lie in the xy-plane

# defining screen
sx = np.linspace(-10, 10, 500)
sy = np.linspace(-10, 10, 500)
SX, SY = np.meshgrid(sx, sy)

# defining wave properties
wavelength = 1
k = 2 * np.pi / wavelength

# calculating the fields

E_total1 = np.zeros_like(SX, dtype=complex)
SZ = 0  # screen is a plane specified height
for x in px:
    for y in py:
        # distances from point source to every point on the screen
        r = np.sqrt((SX - x)**2 + (SY - y)**2 + (SZ - pz)**2)
        E_total1 += np.exp(1j * k * r) / r

E_total2 = np.zeros_like(SX, dtype=complex)
SZ = 10  # screen is a plane specified height
for x in px:
    for y in py:
        # distances from point source to every point on the screen
        r = np.sqrt((SX - x)**2 + (SY - y)**2 + (SZ - pz)**2)
        E_total2 += np.exp(1j * k * r) / r

E_total3 = np.zeros_like(SX, dtype=complex)
SZ = 20  # screen is a plane specified height
for x in px:
    for y in py:
        # distances from point source to every point on the screen
        r = np.sqrt((SX - x)**2 + (SY - y)**2 + (SZ - pz)**2)
        E_total3 += np.exp(1j * k * r) / r
```

```python
plt.figure(figsize=(16, 6))

plt.subplot(1, 3, 1)
# Plot the real part of the field
plt.contourf(SX, SY, np.abs(E_total1), levels=50, cmap='viridis')
plt.title('Real part of E at distance of 0')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('square')

plt.subplot(1, 3, 2)
# Plot the real part of the field
plt.contourf(SX, SY, np.abs(E_total2), levels=50, cmap='viridis')
plt.title('Real part of E at distance of 10')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('square')

plt.subplot(1, 3, 3)
# Plot the real part of the field
plt.contourf(SX, SY, np.abs(E_total3), levels=50, cmap='viridis')
plt.title('Real part of E at distance of 20')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('square')
```
