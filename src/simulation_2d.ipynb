```python
from dmd import Dmd2d, save_surface
from simulation import Simulation2d
import numpy as np
import matplotlib.pyplot as plt
```

```python
dmd_test = Dmd2d(12, 1, .1, 5, 5, 10, 10)
sim_test = Simulation2d(dmd_test, 0, 0, 1e-1, (100, 100), 10, "spherical")

def plot_field_and_dmd(sim: Simulation2d):
    # Plotting
    plt.figure(figsize=(12, 6))
    
    surface = dmd_test.get_surface_view(10)
    # Plot the surface of the dmd
    plt.subplot(1, 2, 1)
    plt.imshow(surface, cmap='viridis', origin='lower')
    plt.colorbar(label="height")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("dmd from above")

    E_total = sim.get_E_total()
    # Plot the real part of the total reflected field
    plt.subplot(1, 2, 2)
    plt.contourf(sim.X, sim.Y, np.log(np.abs(E_total)), levels=50, cmap='viridis')
    plt.title('abs(E_total)')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.tight_layout()
    plt.show()

plot_field_and_dmd(sim_test)
```
