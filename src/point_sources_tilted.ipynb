```python
import numpy as np
import matplotlib.pyplot as plt
```

# Looking from above

```python
# defining screen
sx = np.linspace(-10, 10, 500)
sy = np.linspace(-10, 10, 500)
SX, SY = np.meshgrid(sx, sy)
SZ = 0  # screen is a plane specified height

# defining wave properties
wavelength = 1
k = 2 * np.pi / wavelength

# defining point sources
nr_sources_x = 10
nr_sources_y = 10
mirror_size = 10

# tilt point sources
tilt_angle_deg = 12
tilt_angle_rad = np.deg2rad(tilt_angle_deg)
z = 0

E_total_top = np.zeros_like(SX, dtype=complex)
for i in range(nr_sources_x):
    x = -mirror_size/2 + mirror_size/nr_sources_x * i
    for j in range(nr_sources_y):
        y = -mirror_size/2 + mirror_size/nr_sources_y * j
        x_tilted = x
        y_tilted = y*np.cos(tilt_angle_rad) - z*np.sin(tilt_angle_rad)
        z_tilted = y*np.sin(tilt_angle_rad) + z*np.sin(tilt_angle_rad)
        r = np.sqrt((SX - x_tilted)**2 + (SY - y_tilted)**2 + (SZ - z_tilted)**2)
        E_total_top += np.exp(1j * k * r) / r
```

```python
plt.contourf(SX, SY, np.log(np.abs(E_total_top)), levels=50, cmap='viridis')
plt.title('Real part of E')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('square')
```
